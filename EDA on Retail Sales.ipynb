```python
# Load the retail sales data from a CSV file into a pandas DataFrame
df = pd.read_csv('retail_sales.csv')
```

```python
# Calculate the Total Sales for each order and add it as a new column to the DataFrame
df['Total_Sales'] = df['unit_price'] * df['quantity']
```

```python
# Calculate the Total Cost for each order and add it as a new column to the DataFrame
df['Total Cost'] = df['unit_cost'] * df['quantity']
```

```python
# Calculate the Profit for each order (Total Sales - Total Cost) and add it as a new column
df['Profit'] = df['Total_Sales'] - df['Total Cost']
```

### Columns Description
##### order_id : uniques_id

##### order_date: The date on which the order was placed.

##### customer_id: A unique identifier for the customer who placed the order.

##### branch: The specific store location or branch where the order was processed or fulfilled.

##### product_category: The category or type of the product sold (e.g., electronics, clothing).

##### product_model: The specific model or version of the product sold.

##### unit_cost: The cost price per unit of the product (how much the company paid for each item).

##### unit_price: The selling price per unit of the product (how much the customer paid per item).

##### quantity: The number of units of the product purchased in the order.

##### shipping_cost: The cost charged for delivering the product to the customer.

##### payment_method: The method used by the customer to pay for the order (e.g., credit card, cash, online payment).

##### sales_channel: The channel through which the sale was made (e.g., online store, physical store, mobile app).

##### discount: Any discount applied to the order or product, usually represented as an amount or percentage.

##### return_status: Indicates if the product was returned after purchase (e.g., returned, not returned).

##### loyalty: Indicates if the customer is part of a loyalty program or has loyalty points applied to the purchase.

#### Numerical Columns
## Numerical Columns
unit_cost: Cost price per unit (numeric value).  
unit_price: Selling price per unit (numeric value).  
quantity: Number of units purchased (numeric value).  
shipping_cost: Cost charged for delivery (numeric value).  
discount: Discount applied (numeric value).  

## Categorical Columns
branch: Store location or branch (category names).  
product_category: Category/type of product (e.g., electronics, clothing).  
product_model: Specific model/version of product (categorical identifier).  
payment_method: Method of payment used (e.g., credit card, cash).  
sales_channel: Sales channel (e.g., online, in-store).  
return_status: Indicates if product was returned (e.g., returned, not returned).  
loyalty: Indicates loyalty program status (e.g., yes/no).  

## Mixed Columns
order_id: Unique order identifier, mostly alphanumeric, not for calculations.  
order_date: Date of order, can be treated as datetime type (special format).  
customer_id: Unique customer identifier, often alphanumeric, used as key/reference.



### Steps of doing Univariate Analysis on Numerical columns

- **Descriptive Statistics**: Compute basic summary statistics for the column, such as mean, median, mode, standard deviation, range, and quartiles. These statistics give a general understanding of the distribution of the data and can help identify skewness or outliers.

- **Visualizations**: Create visualizations to explore the distribution of the data. Some common visualizations for numerical data include histograms, box plots, and density plots. These visualizations provide a visual representation of the distribution of the data and can help identify skewness an outliers.

- **Identifying Outliers**: Identify and examine any outliers in the data. Outliers can be identified using visualizations. It is important to determine whether the outliers are due to measurement errors, data entry errors, or legitimate differences in the data, and to decide whether to include or exclude them from the analysis.

- **Skewness**: Check for skewness in the data and consider transforming the data or using robust statistical methods that are less sensitive to skewness, if necessary.

- **Conclusion**: Summarize the findings of the EDA and make decisions about how to proceed with further analysis.

# Categorical-Categorical Analysis

```python
# Create a contingency table and heatmap to visualize the distribution of payment methods by sales channel
# Create a contingency table
contingency_table_sales_payment = pd.crosstab(df['sales_channel'], df['payment_method'])

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(contingency_table_sales_payment, annot=True, fmt='d', cmap='viridis')
plt.title('Payment Method Distribution by Sales Channel')
plt.xlabel('Payment Method')
plt.ylabel('Sales Channel')
plt.show()
```

```python
# Create a contingency table and heatmap to visualize the distribution of return status by product category
# Create a contingency table
contingency_table_category_return = pd.crosstab(df['product_category'], df['return_status'])

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table_category_return, annot=True, fmt='d', cmap='viridis')
plt.title('Return Status Distribution by Product Category')
plt.xlabel('Return Status')
plt.ylabel('Product Category')
plt.show()
```

```python
# Create a contingency table and heatmap to visualize the distribution of sales channels by branch
# Create a contingency table
contingency_table_branch_sales = pd.crosstab(df['branch'], df['sales_channel'])

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(contingency_table_branch_sales, annot=True, fmt='d', cmap='viridis')
plt.title('Sales Channel Distribution by Branch')
plt.xlabel('Sales Channel')
plt.ylabel('Branch')
plt.show()
```

```python
# Create a contingency table and heatmap to visualize the distribution of product categories by branch
# Create a contingency table
contingency_table_branch_payment = pd.crosstab(df['branch'], df['product_category'])

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(contingency_table_branch_payment, annot=True, fmt='d', cmap='viridis')
plt.title(' product_category Distribution by Branch')
plt.xlabel('product_category')
plt.ylabel('Branch')
plt.show()
```
